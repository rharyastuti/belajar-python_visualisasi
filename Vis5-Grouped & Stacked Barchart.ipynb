```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

Expected Output :



```python
## grouping data
grouped = df.groupby('Genre')['NA_Sales','EU_Sales','JP_Sales','Other_Sales'].mean()
grouped
```

<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
## menggunakan style 'seaborn'
plt.style.use('ggplot')
```

```python
## membuat objek figure
fig, ax = plt.subplots(figsize=(20,10))

## buat peubah untuk barplot
sumbu_x = grouped.columns
y0 = grouped.iloc[0]
y1 = grouped.iloc[1]
y2 = grouped.iloc[2]
y3 = grouped.iloc[3]
y4 = grouped.iloc[4]
y5 = grouped.iloc[5]
y6 = grouped.iloc[6]
y7 = grouped.iloc[7]
y8 = grouped.iloc[8]
y9 = grouped.iloc[9]
y10 = grouped.iloc[10]
y11 = grouped.iloc[11]

## x-values of the bars
x = np.arange(len(sumbu_x))

## width of the bars
w = 0.05

## visualisasi dengan barplot
ax.bar(x, y0, width=w, label=grouped.index.values[0])
ax.bar(x+w, y1, width=w, label=grouped.index.values[1])
ax.bar(x+2*w, y2, width=w, label=grouped.index.values[2])
ax.bar(x+3*w, y3, width=w, label=grouped.index.values[3])
ax.bar(x+4*w, y4, width=w, label=grouped.index.values[4])
ax.bar(x+5*w, y5, width=w, label=grouped.index.values[5])
ax.bar(x+6*w, y6, width=w, label=grouped.index.values[6])
ax.bar(x+7*w, y7, width=w, label=grouped.index.values[7])
ax.bar(x+8*w, y8, width=w, label=grouped.index.values[8])
ax.bar(x+9*w, y9, width=w, label=grouped.index.values[9])
ax.bar(x+10*w, y10, width=w, label=grouped.index.values[10])
ax.bar(x+11*w, y11, width=w, label=grouped.index.values[11])

## memberi keterangan pada grafik
plt.xticks(np.arange(0.25, 4.25, 1), ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'])
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games By Genre', fontweight='bold')

## memberi legenda
ax.legend(loc='upper right', bbox_to_anchor=(0.75,0.75,0.3,0.3), fancybox=True, facecolor='inherit').get_frame()
```

```python
grouped.mean(axis=1).sort_values(ascending=True)
```

## Interpretasi 'Grouped bar chart of Mean Sales Video Games by Genre'

Berdasarkan grouped bar chart tersebut, secara general rata-rata penjualan video games yang paling tinggi berada di North America (NA_sales), dimana genre tertinggi yang terjual adalah <b>platform</b> dan disusul dengan <b>shooter</b>. Genre <b>platform</b> berhasil menduduki peringkat pertama rata-rata penjualan di North America, Europe, dan Japan. Lalu secara general diikuti oleh genre <b>shooter</b> dan <b>role-playing</b>. Adapun video game yang menduduki peringkat terendah adalah genre <b>adventure</b> dan <b>strategy</b>.


<h2>Perbandingan Kuantitatif Barplot : Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
## membuat objek figure
fig, ax = plt.subplots(figsize=(20,10))

## buat peubah untuk barplot
sumbu_x = grouped.columns
y0 = grouped.iloc[0]
y1 = grouped.iloc[1]
y2 = grouped.iloc[2]
y3 = grouped.iloc[3]
y4 = grouped.iloc[4]
y5 = grouped.iloc[5]
y6 = grouped.iloc[6]
y7 = grouped.iloc[7]
y8 = grouped.iloc[8]
y9 = grouped.iloc[9]
y10 = grouped.iloc[10]
y11 = grouped.iloc[11]

## x-values of the bars
x = np.arange(len(sumbu_x))

## width of the bars
w = 0.75

## visualisasi dengan barplot
ax.bar(x, y0, width=w, label=grouped.index.values[0])
ax.bar(x, y1, width=w, bottom=y0, label=grouped.index.values[1])
ax.bar(x, y2, width=w, bottom=y0+y1, label=grouped.index.values[2])
ax.bar(x, y3, width=w, bottom=y0+y1+y2, label=grouped.index.values[3])
ax.bar(x, y4, width=w, bottom=y0+y1+y2+y3, label=grouped.index.values[4])
ax.bar(x, y5, width=w, bottom=y0+y1+y2+y3+y4, label=grouped.index.values[5])
ax.bar(x, y6, width=w, bottom=y0+y1+y2+y3+y4+y5, label=grouped.index.values[6])
ax.bar(x, y7, width=w, bottom=y0+y1+y2+y3+y4+y5+y6, label=grouped.index.values[7])
ax.bar(x, y8, width=w, bottom=y0+y1+y2+y3+y4+y5+y6+y7, label=grouped.index.values[8])
ax.bar(x, y9, width=w, bottom=y0+y1+y2+y3+y4+y5+y6+y7+y8, label=grouped.index.values[9])
ax.bar(x, y10, width=w, bottom=y0+y1+y2+y3+y4+y5+y6+y7+y8+y9, label=grouped.index.values[10])
ax.bar(x, y11, width=w, bottom=y0+y1+y2+y3+y4+y5+y6+y7+y8+y9+y10, label=grouped.index.values[11])

## memberi keterangan pada grafik
plt.xticks(np.arange(0, 4.75, 1), ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'])
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games By Genre', fontweight='bold')

## memberi legenda
ax.legend(loc='upper right', bbox_to_anchor=(0.75,0.75,0.3,0.3), fancybox=True, facecolor='inherit').get_frame()
```

## Interpretasi 'Stacked bar chart of Mean Sales Video Games by Genre'

Seperti halnya informasi yang didapatkan melalui grouped bar chart, rata-rata penjualan games yang paling tinggi berada di North America. Informasi tersebut lebih mudah didapat melalui <i>stacked bar chart</i> karena memberikan ukuran yang paling tinggi. Masih di North America, genre tertinggi yang terjual adalah <b>platform</b> dan disusul dengan <b>shooter</b>, yang dapat dilihat ketebalan lapisan (bayangkan roti lapis) atau tinggi stacked-bar. Adapun video game yang menduduki peringkat terendah adalah genre <b>adventure</b> dan <b>strategy</b>. Hal ini dapat dilihat bahwa di setiap negara/ negara bagian, ketebalan lapisan kedua genre tersebut adalah yang konsisten tipis.

Secara umum <i>stacked bar chart</i> memberikan informasi terkait proporsi antar kategori pada setiap peubah X (dalam hal ini negara/ negara bagian) yang ingin dibandingkan.
