<h2>Tugas Harian No 2 : Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
## expected output 1
## menampilkan objek visual
plt.plot()
plt.show()
```

```python
## expected output 2
## menambahkan (plotting) data ke dalam objek axes
plt.plot(x, y)

## menampilkan plot data
plt.show()
```

<h2>Tugas Harian No 3 : Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
## melihat style plotting yang disediakan matplotlib
plt.style.available
```

```python
## menggunakan style 'seaborn'
plt.style.use('seaborn')
```

```python
## plotting
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)

## menampilkan plot
plt.show
```

Expected Output :



```python
## menyimpan grafik as picture
plt.savefig('Week 3-Seaborn.png', bbox_inches='tight')
```
