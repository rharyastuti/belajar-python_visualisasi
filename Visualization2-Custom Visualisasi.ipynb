Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note : warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

Expected Result :



```python
## membuat komponen figure dan axis
fig, ax = plt.subplots()

## plotting cos(x)
ax.plot(x, cos, color='red', label='Cosx')

## plotting 3*cos(x)
ax.plot(x, cos3, color='blue', linestyle='dashdot', label='3 Cosx')

## plotting sin(x)
ax.plot(x, sin, color='green', linestyle='dashed', label='Sin(x-90)')

## plotting 2*sin(x)
ax.plot(x, sin2, color='yellow', linestyle='None', marker='x', label='2 Sinx')

## memberi label sumbu dan judul grafik
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')

## memberi legenda
ax.legend(loc='center')

## menampilkan plot
plt.show()
```
